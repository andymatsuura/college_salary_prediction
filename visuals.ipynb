```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cards = pd.read_csv("resources/cards.csv")
```

```python
top_schools = cards.sort_values(by="mn_earn_wne_p6", ascending=False).head(12)
```

```python
top_schools.plot(kind='bar', x="INSTNM", y="mn_earn_wne_p6", label='Average Annual Income')
plt.title('Bar Plot of Top 12 Schools') 
```

```python
bottom_schools = cards.sort_values(by="mn_earn_wne_p6", ascending=True).head(12)
```

```python
bottom_schools.plot(kind='bar', x="INSTNM", y="mn_earn_wne_p6", label='Average Annual Income')
plt.title('Bar Plot of Bottom 12 Schools') 
```

```python
avg_earn_state = cards.groupby('st_fips')['mn_earn_wne_p6'].mean()
top_avg_earn_state = avg_earn_state.sort_values(ascending=False).head(12)
top_avg_earn_state.plot(kind='bar')
```

```python
low_avg_earn_state = avg_earn_state.sort_values(ascending=False).tail(12)
low_avg_earn_state.plot(kind='bar')
```

```python
avg_debt = cards["DEBT_MDN"].mean()
avg_debt
```

```python
top_debt = cards.nlargest(10, 'DEBT_MDN')
top_debt.plot(kind="bar", x="INSTNM", y="DEBT_MDN")
```

```python
less_debt = cards.nsmallest(10, 'DEBT_MDN')
less_debt.plot(kind="bar", x="INSTNM", y="DEBT_MDN")
```

```python
# Get the top 10 rows based on mn_earn_wne_p6
top_10 = cards.nlargest(10, 'mn_earn_wne_p6')

# Extract data for plotting
x_labels = top_10['INSTNM']
mn_earnings = top_10['mn_earn_wne_p6']
debt_median = top_10['DEBT_MDN']

# Assign numeric positions to the x-axis labels
x = np.arange(len(x_labels))
bar_width = 0.35

# Plotting the bar graph
fig, ax = plt.subplots(figsize=(14, 8))

# Create the bar plots
bar1 = ax.bar(x - bar_width/2, mn_earnings, bar_width, label='Average Earnings')
bar2 = ax.bar(x + bar_width/2, debt_median, bar_width, label='Median Debt')

# Add labels, title, and legend
ax.set_xlabel('Institution')
ax.set_ylabel('Value')
ax.set_title('Top 10 Institutions by Average Earnings and Median Debt')
ax.set_xticks(x)
ax.set_xticklabels(x_labels, rotation=45, ha='right')
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()

```
